```python
def prime_gen(a):
    prime_list=[]                    
    for i in range(a): 
        prime_flag='Y' 
        if i==0 or i==1: 
            prime_flag='N' 
        else: 
            for j in range(i): 
                if j!=0 and j!=1: 
                    if i%j==0: 
                        prime_flag="N" 
        if prime_flag=='Y': 
              prime_list.append(i) 
    yield prime_list 
    
    
def find_factors(a): 
#    print(a)
    fact_list=[] 
    l=[]
    prime_list = prime_gen(a) 
 
    for i in prime_list: 
#        print(f'test {i}')
        l=i
    f='Y'
    while f=='Y': 
        a,fact_list=facts(a,l,fact_list) 
#        print('a: ', a)
        if a==1: 
            f='N'
    return fact_list 


def facts(n,l,fact_list1): 

    for i in l:
#        print(type(i),i)
        if n%i==0: 
            fact_list1.append(i) 
            n=n//i 
#            if n==1:
#                 print("n is one")
            break
    return n,fact_list1 
#r=find_factors(12)
#print(f'rprint',r)

def find_lcm(fact_lst1,fact_lst2):
    
    for i in fact_lst1:
        if i in fact_lst2:
            j=fact_lst2.index(i)
            fact_lst2.pop(j)
#    print(f'fact_list1 {fact_list1} fact_list2: {fact_list2}')
    lcm_list=fact_lst1 + fact_lst2
#    print(f'LCM factors list {lcm_list}')
    lcm=1
    for i in lcm_list:
        lcm=lcm*i
    return lcm


def find_hcf(fct_lst1,fct_lst2):
    hcf_list=[]
    for i in fct_lst1:
        if i in fct_lst2:
            hcf_list.append(i)
            j=fct_lst2.index(i)
            fct_lst2.pop(j)
        
#    print(f'fact_list1 {fact_list1} fact_list2: {fact_list2}')
#    print(f'HCF factors list {hcf_list}')
    hcf=1
    for i in hcf_list:
        hcf=hcf*i
    return hcf

a=int(input("Enter the number1: \n"))        
b=int(input("Enter the number2: \n")) 
fact_list1=find_factors(a) 

print(f'Factors of {a} are : {fact_list1}')
fact_list2=find_factors(b) 
print(f'Factors of {b} are : {fact_list2}')
import copy

lcm_list1=copy.deepcopy(fact_list1)
hcf_list1=copy.deepcopy(fact_list1)
lcm_list2=copy.deepcopy(fact_list2)
hcf_list2=copy.deepcopy(fact_list2)
lcm=find_lcm(lcm_list1,lcm_list2)
#print(f'fact_list1 {fact_list1}, fact_list2 {fact_list2}')
print(f'LCM of {a} and {b} is {lcm}')
hcf=find_hcf(hcf_list1,hcf_list2)
print(f'HCF of {a} and {b} is {hcf}')
```
